```python
train1['Answer'] = train1['Short_Answer']+'. '+train1['Long_Answer']
train1=train1.drop(['Long_Answer','Short_Answer'], axis=1)

test['Answer'] = test['Short_Answer']+'. '+test['Long_Answer']
test=test.drop(['Long_Answer','Short_Answer'], axis=1)

train1['User_ID']=train1['User_ID']+240
train1
```

```python
train=pd.concat([train0, train1], ignore_index=True)
train
```

```python
train['Gender']=train['Gender'].replace('female', '0')
train['Gender']=train['Gender'].replace('male', '1')

test['Gender']=test['Gender'].replace('female', '0')
test['Gender']=test['Gender'].replace('male', '1')

train
```

```python
train['Gender']=train['Gender'].astype(int)
test['Gender']=test['Gender'].astype(int)

def remove_last_space(string):
    if string[-1] == ' ':
        return string[:-1]
    else:
        return string

# Apply the function to the desired column
train['Answer'] = train['Answer'].apply(lambda x: remove_last_space(x))
test['Answer'] = test['Answer'].apply(lambda x: remove_last_space(x))
```
